```python
# Step 1: Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress, pearsonr

# Step 2: Input your dataset
data = {
    "Density": [1126.98, 1125.36, 1125.88, 1126.78, 1124.00, 1124.00, 1121.00],
    "CRM_TS": [49.5, 48.9, 48.0, 50.8, 48.8, 49.7, 47.5],
    "CCIC_TS": [50.9, 50.4, 50.1, 51.6, 50.4, 50.1, 48.7]
}
df = pd.DataFrame(data)

# Step 3: Correlation and regression - Density vs CRM
r_crm, _ = pearsonr(df["Density"], df["CRM_TS"])
reg_crm = linregress(df["Density"], df["CRM_TS"])

# Step 4: Correlation and regression - Density vs CCIC
r_ccic, _ = pearsonr(df["Density"], df["CCIC_TS"])
reg_ccic = linregress(df["Density"], df["CCIC_TS"])

# Step 5: Print results
print("=== Correlation & Regression ===\n")
print("Density vs CRM Total Solids")
print(f"  Pearson r: {r_crm:.3f}")
print(f"  Regression: y = {reg_crm.slope:.3f}x + {reg_crm.intercept:.2f}")
print(f"  R² = {reg_crm.rvalue**2:.3f}, P-value = {reg_crm.pvalue:.4f}\n")

print("Density vs CCIC Total Solids")
print(f"  Pearson r: {r_ccic:.3f}")
print(f"  Regression: y = {reg_ccic.slope:.3f}x + {reg_ccic.intercept:.2f}")
print(f"  R² = {reg_ccic.rvalue**2:.3f}, P-value = {reg_ccic.pvalue:.4f}")

# Step 6: Plot Density vs CRM TS
plt.figure(figsize=(8, 5))
plt.scatter(df["Density"], df["CRM_TS"], color="blue", label="CRM TS")
plt.plot(df["Density"], reg_crm.intercept + reg_crm.slope * df["Density"], color="black", label="Fit line")
plt.title("Density vs CRM Total Solids")
plt.xlabel("Density")
plt.ylabel("CRM Total Solids (%)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Step 7: Plot Density vs CCIC TS
plt.figure(figsize=(8, 5))
plt.scatter(df["Density"], df["CCIC_TS"], color="green", label="CCIC TS")
plt.plot(df["Density"], reg_ccic.intercept + reg_ccic.slope * df["Density"], color="black", label="Fit line")
plt.title("Density vs CCIC Total Solids")
plt.xlabel("Density")
plt.ylabel("CRM Total Solids (%)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 📓 Jupyter Notebook Code for Provided Data

import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

# 1️⃣ Create dataframe from your table
data = {
    'Sample_ID': ['S1','S2','S3','S4','S5','S6','S7'],
    'Density': [1126.98,1125.36,1125.88,1126.78,1124.0,1124.0,1121.0],
    'CRM_Total_Solids': [49.5,48.9,48.0,50.8,48.8,49.7,47.5],
    'CCIC_Total_Solids': [50.9,50.4,50.1,51.6,50.4,50.1,48.7]
}
df = pd.DataFrame(data)

# 2️⃣ Perform linear regression: CRM
slope_crm, intercept_crm, r_crm, p_crm, stderr_crm = linregress(
    df['Density'], df['CRM_Total_Solids']
)

# 3️⃣ Perform linear regression: CCIC
slope_ccic, intercept_ccic, r_ccic, p_ccic, stderr_ccic = linregress(
    df['Density'], df['CCIC_Total_Solids']
)

# 4️⃣ Add predicted values
df['CRM_Predicted'] = df['Density'] * slope_crm + intercept_crm
df['CCIC_Predicted'] = df['Density'] * slope_ccic + intercept_ccic

# 5️⃣ Plot: CRM
plt.figure(figsize=(7,5))
plt.scatter(df['Density'], df['CRM_Total_Solids'], label="CRM Data", color='blue')
plt.plot(df['Density'], df['CRM_Predicted'], color='black',
         label=f"Fit: y={slope_crm:.3f}x+{intercept_crm:.1f}\nR²={r_crm**2:.3f}")
plt.xlabel("Density")
plt.ylabel("CRM Total Solids")
plt.title("Density vs CRM Total Solids Regression for 310 recepie")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# 6️⃣ Plot: CCIC
plt.figure(figsize=(7,5))
plt.scatter(df['Density'], df['CCIC_Total_Solids'], label="CCIC Data", color='green')
plt.plot(df['Density'], df['CCIC_Predicted'], color='black',
         label=f"Fit: y={slope_ccic:.3f}x+{intercept_ccic:.1f}\nR²={r_ccic**2:.3f}")
plt.xlabel("Density")
plt.ylabel("CCIC Total Solids")
plt.title("Density vs CCIC Total Solids Regression for 310 recepie")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# 7️⃣ Print regression equations and R²
print("CRM Regression:")
print(f"y = {slope_crm:.3f} * Density + {intercept_crm:.1f}")
print(f"R² = {r_crm**2:.3f}\n")

print("CCIC Regression:")
print(f"y = {slope_ccic:.3f} * Density + {intercept_ccic:.1f}")
print(f"R² = {r_ccic**2:.3f}")
```
